```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cleaned_e-commerce_dataset.csv")
furniture_df = df[df['Category'] == 'Furniture']
```

```python
kpis = {
    "Total Furniture Sales": furniture_df['Amount'].sum(),
    "Total Furniture Profit": furniture_df['Profit'].sum(),
    "Average Order Value": furniture_df['Amount'].mean(),
    "Average Profit Margin (%)": furniture_df['Profit_Margin'].mean()
}

kpis
```

The furniture category demonstrates high revenue potential with strong profit margins, making it a key focus area for business growth.

```python
top_subcats = furniture_df.groupby('Sub-Category')['Amount'].sum().sort_values(ascending=False)

top_subcats.plot(kind='bar', figsize=(8,5))
plt.title("Top Furniture Sub-Categories by Revenue")
plt.ylabel("Sales Amount")
plt.show()
```

```python
profit_comparison = furniture_df.groupby('Sub-Category')['Profit_Margin'].mean()

profit_comparison.plot(kind='bar', figsize=(8,5))
plt.title("Average Profit Margin by Furniture Sub-Category")
plt.ylabel("Profit Margin (%)")
plt.show()
```

```python
payment_impact = furniture_df.groupby('PaymentMode')['Amount'].sum()

payment_impact.plot(kind='pie', autopct='%1.1f%%', figsize=(6,6))
plt.title("Furniture Sales Distribution by Payment Mode")
plt.ylabel("")
plt.show()
```

```python
sns.regplot(data=furniture_df, x='Quantity', y='Profit')
plt.title("Furniture Quantity vs Profit Trend")
plt.show()
```

Profit does not increase linearly with quantity, reinforcing the importance of pricing and cost control.

Business Insights
- Furniture is a high-value category with strong revenue contribution.
- EMI is the dominant payment mode, indicating customer preference for installment-based purchases.
- Chairs and Bookcases are the best-performing sub-categories in terms of sales.
- High profit margins are not always linked to high sales volumes.
- Optimizing pricing strategies can significantly improve profitability.

Business Recommendations
- Promote high-margin furniture sub-categories through targeted campaigns.
- Strengthen EMI and credit card offers to boost furniture sales.
- Focus on value-based pricing rather than quantity-based discounts.
- Optimize inventory for top-performing sub-categories like Chairs.
- Use profit margin analysis to guide future product launches.
