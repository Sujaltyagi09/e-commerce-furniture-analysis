```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cleaned_e-commerce_dataset.csv")

furniture_df = df[df['Category'] == 'Furniture']
furniture_df.head()
```

```python
total_furniture_sales = furniture_df['Amount'].sum()
total_furniture_profit = furniture_df['Profit'].sum()
avg_order_value = furniture_df['Amount'].mean()
avg_profit_margin = furniture_df['Profit_Margin'].mean()

total_furniture_sales, total_furniture_profit, avg_order_value, avg_profit_margin
```

Furniture category generates high order value with healthy average profit margins, indicating premium product behavior.

```python
sub_category_perf = furniture_df.groupby('Sub-Category')[['Amount','Profit','Quantity']].sum()
sub_category_perf
```

```python
sub_category_perf[['Amount','Profit']].plot(kind='bar', figsize=(9,5))
plt.title("Furniture Sub-Category Sales & Profit")
plt.ylabel("Value")
plt.show()
```

Certain sub-categories contribute more to revenue, while others deliver better profit efficiency.

```python
sns.scatterplot(data=furniture_df, x='Quantity', y='Profit')
plt.title("Furniture Quantity vs Profit")
plt.show()
```

Bulk furniture purchases do not necessarily result in higher profit due to discounting and logistics costs.

```python
furniture_payment = furniture_df.groupby('PaymentMode')[['Amount','Profit']].sum()
furniture_payment
```

```python
furniture_payment = furniture_df.groupby('PaymentMode')[['Amount','Profit']].sum()
furniture_payment
```

```python
profit_margin_sub = furniture_df.groupby('Sub-Category')['Profit_Margin'].mean()
profit_margin_sub
```

```python
profit_margin_sub.plot(kind='bar', figsize=(8,5))
plt.title("Average Profit Margin by Furniture Sub-Category")
plt.ylabel("Profit Margin (%)")
plt.show()
```

Sub-categories with lower sales volume may still offer higher profit margins, making them strategically important.

- Furniture orders have higher average order value compared to other categories.
- EMI dominates furniture transactions, indicating premium purchasing behavior.
- Chairs and Bookcases are major contributors to revenue and quantity sold.
- Profit margin analysis highlights opportunities for optimizing pricing strategies.
= Not all high-quantity orders are profitable, emphasizing margin-based decision-making.
