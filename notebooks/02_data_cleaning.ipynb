```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("e-commerce_dataset.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['Amount'] = pd.to_numeric(df['Amount'], errors='coerce')
df['Profit'] = pd.to_numeric(df['Profit'], errors='coerce')
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
```

```python
df.duplicated().sum()
df.drop_duplicates(inplace=True)
```

```python
df['Category'] = df['Category'].str.strip().str.title()
df['Sub-Category'] = df['Sub-Category'].str.strip().str.title()
df['PaymentMode'] = df['PaymentMode'].str.strip().str.upper()
```

```python
df['Profit_Margin'] = (df['Profit'] / df['Amount']) * 100
```

```python
furniture_df = df[df['Category'] == 'Furniture']
furniture_df.head()
```

```python
furniture_df.describe()
```

```python
df.to_csv("cleaned_e-commerce_dataset.csv", index=False)
```

- The dataset has been cleaned and standardized for analysis.
- Numerical columns are correctly formatted.
- Duplicate records have been removed.
- A new feature, Profit Margin, has been created for deeper profitability analysis.
- Furniture category data is prepared for focused exploration.
