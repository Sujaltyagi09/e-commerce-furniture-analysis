```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cleaned_e-commerce_dataset.csv")
df.head()
```

```python
total_sales = df['Amount'].sum()
total_profit = df['Profit'].sum()

total_sales, total_profit
```

The dataset represents significant e-commerce activity with substantial total sales and profit, indicating diverse purchasing behavior across categories.

```python
category_analysis = df.groupby('Category')[['Amount','Profit']].sum()
category_analysis
```

Furniture contributes a considerable portion of revenue and profit, making it a strong candidate for focused analysis.

```python
category_analysis.plot(kind='bar', figsize=(8,5))
plt.title("Category-wise Sales and Profit")
plt.ylabel("Amount")
plt.show()
```

```python
furniture_df = df[df['Category'] == 'Furniture']

furniture_sub = furniture_df.groupby('Sub-Category')[['Amount','Profit']].sum()
furniture_sub
```

Different furniture sub-categories show varying revenue and profitability patterns.

```python
sns.scatterplot(data=df, x='Quantity', y='Profit')
plt.title("Quantity vs Profit")
plt.show()
```

Higher quantity orders do not always result in higher profit, highlighting the importance of pricing and margins.

```python
payment_analysis = df.groupby('PaymentMode')[['Amount','Profit']].sum()
payment_analysis
```

```python
payment_analysis.plot(kind='bar', figsize=(8,5))
plt.title("Payment Mode vs Sales and Profit")
plt.show()
```

EMI and Credit Card payments are dominant for higher-value purchases, especially in furniture orders.

```python
sns.histplot(df['Profit_Margin'], bins=20, kde=True)
plt.title("Profit Margin Distribution")
plt.show()
```

Most orders fall within a moderate profit margin range, with few high-margin outliers.

- Furniture is a major revenue-generating category.
- Chairs and Bookcases show strong performance among furniture products.
- Payment mode significantly influences purchasing behavior.
- Quantity alone is not a reliable indicator of profitability.
- Profit margin analysis helps identify high-value orders.
